```python
#initialize elastic

from elasticsearch import Elasticsearch
from elasticsearch.helpers import bulk


es = Elasticsearch(
    [{'host': 'localhost', 'port': 9200, 'scheme': 'https'}],
    basic_auth=('elastic', 'datascientest'), 
    request_timeout=600,
    verify_certs=True,  # Check SSL certificats
    ca_certs='../ca/ca.crt'  # Specify CA certificat path

)

```

```python
#test elastic connectivity
if es.ping():
    print("Connection successful")
else:
    print("Connection failed")
```

```python
#Create index in needed
if not es.indices.exists(index="bigdata-jobs"):
   es.indices.create(index="bigdata-jobs")
```

```python
#Create index in needed
if not es.indices.exists(index="bigdata-companies"):
   es.indices.create(index="bigdata-companies")
```

```python
#write jobs on elastic
import pandas as pd
import json

jobs_full_json = jobs_full.to_json(orient='records')

json_records = json.loads(jobs_full_json)

index_name = 'bigdata-jobs'
doctype = 'census_record'


action_list = []
for row in json_records:
    record ={
        '_op_type': 'index',
        '_index': index_name,
        '_source': row,
        '_id': row['offer_link']

    }
    action_list.append(record)
bulk(es, action_list)

```

```python
es.indices.refresh(index='bigdata-jobs')
```
