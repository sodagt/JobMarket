```python
from geopy.geocoders import Nominatim
import pycountry
import mpu

# Provide a unique and descriptive user_agent
geolocator = Nominatim(user_agent="my_geolocation_app_v1.0")
location = geolocator.geocode("El Paso, El Paso County")
print(location.address)

print((location.latitude, location.longitude))

latitude = str(location.latitude)
longitude = str(location.longitude)

location_reverse = geolocator.reverse(latitude+","+longitude)
address = location_reverse.raw['address']
print(address)

country = address.get('country', '')
print(country)

city = address.get('city', '')
print(city)

country_code = address.get('country_code', '')
print(country_code)

country_inf = pycountry.countries.get(name=country)
print(country_inf)

currency = pycountry.currencies.get(numeric=country_inf.numeric)
print(currency)
print(currency.name)
print(currency.alpha_3)




#currency = mpu.units.get_currency(country.upper)
#print(currency)

#currency = pycountry.currencies.get(letter=country.upper())
#print(country)

#location = geolocator.geocode("175 5th Avenue NYC")
#print(location.address)
```
