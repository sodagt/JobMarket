```python
from elasticsearch.helpers import bulk
import time

if not es.indices.exists(index="bigdata-companies"):
   es.indices.create(index="bigdata-companies")

# Function to chunk the DataFrame into smaller batches
def chunk_dataframe(df, chunk_size=1000):
    # Yield chunks of the DataFrame
    for start in range(0, df.shape[0], chunk_size):
        yield df.iloc[start:start + chunk_size]

#Write DF in Elasticsearch
def generate_data_to_elk(df: pd.DataFrame):
    for index, row in df.iterrows():
        yield {
            "_index": "bigdata-companies", 
            #"_id": row["id"],  #specify unique value as key
            "_source": row.to_dict(),
        }

def insert_data_elk(df: pd.DataFrame):
    #Insert DataFrame in Elasticsearch
    actions = []  # List to hold batch actions
    for chunk in chunk_dataframe(df):
        for doc in generate_data_to_elk(chunk):
            actions.append(doc)
        try:
            # Convert DataFrame and insert in Elasticsearch
            #success, failed = bulk(es, generate_data_to_elk(df))
            success, failed = bulk(es, actions)
            print(f"Successfully indexed {success} documents")
            print(f"Failed to index {failed} documents.")
            actions.clear() 
            time.sleep(30)


        except Exception as e:
            print(f"Error: {e}")



insert_data_elk(companies_full)

es.indices.refresh(index='bigdata-companies')
```
